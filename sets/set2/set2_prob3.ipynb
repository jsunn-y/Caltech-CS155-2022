```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
def plot(model, alpha_list):
    weights = np.zeros((len(alpha_list), 9))
    for i, alpha in enumerate(alpha_list):
        if model == 'Lasso':
            clf = Lasso(alpha = alpha)
        elif model == 'Ridge':
            clf = Ridge(alpha = alpha)
        clf.fit(X_train, y_train)
        weights[i,:] = clf.coef_
    plt.plot(alpha_list, weights)
    plt.title(model)
    plt.xlabel('Alpha')
    plt.ylabel('Weight')
    plt.legend(np.arange(1,10), loc='upper right')
```

```python
plot('Lasso', np.linspace(0.01, 3, 30))
plt.savefig('lasso.pdf')
```

```python
plot('Ridge', np.linspace(1, 10000, 10000))
plt.savefig('ridge2.pdf')
```
